```python
import turtle
import random
import time
```

```python
# Create screen
sc = turtle.Screen()
sc.title("Pong game")
sc.bgcolor("white")
sc.setup(width=500, height=600)
# Left paddle
left_pad = turtle.Turtle()
left_pad.speed(0)
left_pad.shape("square")
left_pad.color("black")
left_pad.shapesize(stretch_wid=6, stretch_len=1)
left_pad.penup()
left_pad.goto(-197, 0)

balls = []
color = ["blue", "green", "red", "yellow", "purple"]
start = time.time()

def create_ball(color):
    t = turtle.Turtle()
    t.speed(0)
    t.shape("circle")
    t.color(color)
    t.penup()
    t.goto(230, random.randint(-279, 279))
    t.dx = (random.randint(2, 3)+random.random())
    t.dy = -1*(random.randint(2, 3)+random.random())
    balls.append(t)

# Ball of circle shape
create_ball(color[0])
ball_count = 1
hit_count = 0

# Initialize the score
left_player = 0
z = 0

# Displays the score
sketch = turtle.Turtle()
sketch.speed(0)
sketch.color("blue")
sketch.penup()
sketch.hideturtle()
sketch.goto(0, 260)
sketch.write("Score : 0  Time : 60", align="center", font=("Montserrat", 20, "normal"))

#end lose screen
lose = turtle.Turtle()
lose.speed(0)
lose.color("red")
lose.penup()
lose.hideturtle()
lose.goto(0, 0)

#end win screen
win = turtle.Turtle()
win.speed(0)
win.color("green")
win.penup()
win.hideturtle()
win.goto(0, 0)

# Functions to move paddle vertically
def paddleaup():
    y = left_pad.ycor()
    y += 20
    left_pad.sety(y)

def paddleadown():
    y = left_pad.ycor()
    y -= 20
    left_pad.sety(y)

# Keyboard bindings
sc.listen()
sc.onkeypress(paddleaup, "w")
sc.onkeypress(paddleadown, "d")

def update_game():
    global z
    global left_player
    global ball_count
    global hit_count
    turtle.tracer(0, 0)
    for hit_ball in balls:
        hit_ball.setx(hit_ball.xcor() + hit_ball.dx)
        hit_ball.sety(hit_ball.ycor() + hit_ball.dy)

        # Checking borders
        if hit_ball.ycor() > 280 or hit_ball.ycor() < -280:
            hit_ball.dy *= -1
            

        if hit_ball.xcor() > 250:
            hit_ball.dx *= -1

        if (hit_ball.xcor() < -250 and hit_ball.xcor() > -300):
            ball_count -= 1 
            hit_ball.goto(-350, 0)
            

        if (hit_ball.xcor() < -164 and hit_ball.xcor() > -174) and (hit_ball.ycor() < left_pad.ycor() + 69 and hit_ball.ycor() > left_pad.ycor() - 69):
            hit_ball.setx(-160)
            hit_ball.dx *= -1
            hit_count += 1
            left_player += ball_count
            if hit_count % 3 == 0:
                z += 1
                create_ball(color[z%5])
                ball_count += 1
        sketch.clear()
        sketch.write("Balls: {} | Score : {} | Time : {}".format(ball_count, 
                       left_player, 60-(int(time.time() - start))), align="center",
                      font=("Montserrat", 20, "normal"))

        if (ball_count == 0):
            left_pad.color("white")
            sketch.clear()
            lose.write("You Lose! Score: {}".format(left_player), align="center", font=("Montserrat", 30, "normal"))
        if (60 - (int(time.time() - start)) == 0):
            left_pad.color("white")
            sketch.clear()
            win.write("You Win! Score: {}".format(left_player), align="center", font=("Montserrat", 30, "normal"))
            for hit_ball in balls:
                hit_ball.clear()

    sc.update()
    sc.ontimer(update_game, 10)

# Start the game loop
update_game()
sc.mainloop()
```
